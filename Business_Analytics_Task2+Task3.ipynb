2.1.Use simple linear regression to predict CO2 emissions using volume and weight separately, plot their graphs and calculate the r values

```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df=pd.read_csv("que2.csv")
```

```python
x1=df.Volume
x2=df.Weight
y=df.CO2
```

```python
slope, intercept, r, p, std_err = stats.linregress(x1, y)

def myfunc(x1):
  return slope * x1 + intercept

mymodel = list(map(myfunc, x1))

plt.scatter(x1, y)
plt.plot(x1, mymodel)
plt.show()
print(r)
```

```python
slope, intercept, r, p, std_err = stats.linregress(x2, y)

def myfunc(x2):
  return slope * x2 + intercept

mymodel = list(map(myfunc, x2))

plt.scatter(x2, y)
plt.plot(x2, mymodel)
plt.show()
print(r)
```

2.2 Use multiple regression to predict CO2 emission using both weight and volume

```python
X=df[["Volume","Weight"]]
regr = linear_model.LinearRegression()
regr.fit(X, y)
predictedCO2 = regr.predict([[2300, 1300]])
print(predictedCO2)
```
